## Data Cleaning 

Here, we are going to rename some columns as well as modifying records with proper format for further analysis.


```python
df_cal = df_raw.copy()
df_cal.rename(columns={'gender':'Gender'
                       ,'customerID':'CustomerID'
                       ,'Contract':'ContractType'
                       ,'InternetService':'InternetServiceType'
                       ,'tenure':'Tenure'
                      }
              ,inplace=True)


df_cal['Partner'] = df_cal.Partner.map({'Yes':1,'No':0})
df_cal['Dependents'] = df_cal.Dependents.map({'Yes':1,'No':0})

df_cal['PhoneService'] = df_cal.PhoneService.map({'Yes':1,'No':0})
df_cal['MultipleLines'] = df_cal.MultipleLines.map({'Yes':1,'No':0,'No phone service':0})

df_cal['InternetService'] = df_cal.InternetServiceType.map({'DSL':1,'Fiber optic':1,'No':0})
df_cal['OnlineSecurity'] = df_cal.OnlineSecurity.map({'Yes':1,'No':0,'No internet service':0})
df_cal['OnlineBackup'] = df_cal.OnlineBackup.map({'Yes':1,'No':0,'No internet service':0})
df_cal['DeviceProtection'] = df_cal.DeviceProtection.map({'Yes':1,'No':0,'No internet service':0})
df_cal['TechSupport'] = df_cal.TechSupport.map({'Yes':1,'No':0,'No internet service':0})
df_cal['StreamingTV'] = df_cal.StreamingTV.map({'Yes':1,'No':0,'No internet service':0})
df_cal['StreamingMovies'] = df_cal.StreamingMovies.map({'Yes':1,'No':0,'No internet service':0})

df_cal['IsContracted'] = df_cal.ContractType.map({'One year':1,'Two year':1,'Month-to-month':0})
df_cal['PaperlessBilling'] = df_cal.PaperlessBilling.map({'Yes':1,'No':0})
df_cal['Churn'] = df_cal.Churn.map({'Yes':1,'No':0})

df_cal.head()
```

```python
# here we convert the totalchares to numircal value , safely convert numeric strings to float and turn blanks or invalid strings into NaN.
#then replace the NaN value with MonthlyCharged * Tenure 

df_cal['TotalCharges'] =pd.to_numeric(df_cal['TotalCharges'], errors='coerce').fillna(df_cal['MonthlyCharges'] * df_cal['Tenure']).astype(float)
```

```python
# Now we are going to create a feature column called "AnyInternetService" 
#to see if the customer have any internet service at all 
def createAnyInternetservice(row):
    if ((row['OnlineSecurity'] == 1) 
        | (row['OnlineBackup'] == 1) 
        | (row['DeviceProtection'] == 1) 
        | (row['TechSupport'] == 1)
        | (row['StreamingTV'] == 1) 
        | (row['StreamingMovies'] == 1)):
        return 1 
    else:
        return 0

df_cal['AnyInternetService']=df_cal.apply(lambda row : createAnyInternetservice(row), axis= 1  )
df_cal.head()
        
```

This function is designed to analyze and visualize how a categorical variable (like gender, internet service, etc.) is distributed based on customer churn (Yes/No).

🔍 Function Purpose:

To display percentage-based bar plots of a given categorical feature, optionally filtered by another binary feature (like AnyInternetService == 1).

🔧 Function Parameters:
* feature: The categorical variable you want to analyze (e.g., 'Gender', 'InternetService', etc.)

* filterFea: Optional. Another feature used to filter the data first, e.g., 'AnyInternetService'. Only rows where filterFea == 1 will be included if specified.

🔹 If filterFea is provided:

* Filters the data where filterFea == 1.

* Groups by Churn → for each churn group (Yes/No), calculates normalized (percentage) value counts of feature.



```python
def categorical_var_churn_dist(feature,filterFea=''):
    if filterFea != '':
        feature_summary = df_cal[df_cal[filterFea]==1].groupby('Churn')[feature].value_counts(normalize=True).reset_index(name='Percentage')
        print(feature," summary: \n ", feature_summary)
        sns.catplot(hue=feature
                    , y='Percentage'
                    , data = feature_summary
                    , col='Churn'
                    , kind='bar'
                    , alpha = 0.6
                    ,palette='muted')
        
        plt.figure(figsize=(16,8))
        plt.show()
    else:
        feature_summary = churn_summary[feature].value_counts(normalize=True).reset_index(name='Percentage')
        print(feature," summary: \n ", feature_summary)
        sns.catplot(hue=feature
                    , y='Percentage'
                    , data = feature_summary
                    , col='Churn'
                    , kind='bar'
                    , alpha = 0.6
                    ,palette='muted'
                    )
        
        
        plt.figure(figsize=(16,8))
        plt.show()
```

```python
def numerical_distribution(feature):
    fig = plt.figure(figsize=(15,10))
    # "Use a 2-row, 1-column layout, and I'm plotting in position 1 (top)."
    plt.subplot(2,1,1)
    #  draws a Kernel Density Estimate plot (a smoothed histogram).
    ax = sns.kdeplot(df_cal[feature]
                     , color = 'g'
                     , fill = True # makes the area under the curve colored.
                     ,label=feature) # adds a legend label.
   
    plt.title(feature + " Distribution")
    plt.legend()

    
    plt.subplot(2,1,2)  # Now we're on the second subplot (bottom half).
    ax = sns.kdeplot(df_cal.loc[(df_cal['Churn']==1),feature] # Draws the distribution of the feature only for churned customers.
                     , color = 'r'
                     , fill = True
                     , label='Churn')
    plt.legend()

    ax = sns.kdeplot(df_cal.loc[(df_cal['Churn']==0) ,feature] #  draws the distribution for non-churned customers.
                     , color = 'b'
                     , fill = True
                     , label='No chrun')
    
    plt.title( feature + " Distribution: Churn vs No churn")
    plt.legend()
    plt.show()
```

### **K-means clustering**

Here, to see different kind of clusters, we are using K-means cluster method. Before passing data to K-means algorithm, let's normalize Tenure and MonthlyCharges to Standard scale.

You're taking the `Tenure` column and:

- Applying `StandardScaler()` → this **centers** the data around `0` and **scales** it to have **unit variance** (standard deviation = 1).
- `fit_transform()` learns the mean and std, then transforms the data.
- The result is stored in a new column `Tenure_norm`.

**So instead of raw values like `30, 60, 90` months, you get values like `-1.2, 0.3, 1.8`**, depending on how far they are from the mean.

```python
#   STEP 1 STANDAREDIZE THE FEATURES 

import sklearn.preprocessing as prepro 
# normalized tenure and monthlycharges

df_cal['Tenure_norm'] =prepro.StandardScaler().fit_transform(df_cal[['Tenure']])
df_cal['MonthlyCharges_norm'] = prepro.StandardScaler().fit_transform(df_cal[['MonthlyCharges']])
df_cal.head()
```

```python
from sklearn.cluster import KMeans
#   STEP 2 : FILTER THE DATAFRAME 
#---------------------------------------
#Prepare Data (Only churned customers)
#filtering only rows where Churn == 1 and keeping two standardized features.
df_kmean_data = df_cal.loc[df_cal['Churn'] == 1 , ['Tenure_norm' , 'MonthlyCharges_norm']]

#   STEP 3 :FIT multiple KMeans models
#----------------------------------------
#Try different K values from 1 to 9`
k= range(1,10)
#This creates 9 KMeans models with cluster counts 1 through 9.
kmeans = [KMeans(n_clusters= i ) for i in k ]
#Calculate score (inertia) for each K
#.score() returns the negative sum of squared distances between each point and its cluster center (called inertia).
#Smaller score = tighter clusters.
#You're storing those scores for each K.


#   STEP 4 : SCORE /Calculate “inertia” (cluster tightness score)
# ----------------------------------------------------------------

score = [kmeans[i].fit(df_kmean_data).score(df_kmean_data) for i in range(len(kmeans))]

#   STEP 5 :ELBOW / PLOT the elbow

#Plot the Elbow Curve
plt.figure(figsize=(10,6))
plt.plot(k, score)
plt.xlabel("Clusters")
plt.ylabel("Score")
plt.title("Elbow curve", fontsize=15)
```

- The curve typically **drops quickly** at first (more clusters = better fit)
- But at some point, **the improvement slows down** — forming an "elbow" shape.
- That elbow point is usually the **optimal number of clusters**.

## 📌 Summary of What we Just Did

| Step | Action |
| --- | --- |
| 1 | STANDAREDLIZE/ Normalize features so they’re on the same scale |
| 2 | FILTER/ Use only churned customers |
| 3 | FIT/ Try K=1 to K=9 clusters |
| 4 | SCORE/ Fit KMeans and store score (inertia) |
| 5 | ELBOW/ Plot Elbow Curve to decide best K |
| ✅ | **K=3 looks like the best choice** |

## 📌 What we are going to do  — In Plain English

we are going to :

- Clustere churned customers into 3 groups using standardized tenure and charges
- Color each cluster differently
- Visualize the spread and centers of the clusters

This helps us **understand different types of customers who churn**.

For example, clusters might reveal:

- Short-tenure + high-charge users churn the most
- Long-tenure + low-charge users are rare but still churn
- Moderate users behave differently

```python
#Set a Default Cluster Label
df_cal['Cluster'] = -1 #This initializes a new column Cluster with -1 (meaning “not assigned yet”).
# --------------------------------------------------------------------------------------------------
#   Initialize the KMeans Model
#   n_clusters=3: we’re telling KMeans to create 3 clusters.
#   random_state=2: Sets a seed to make the result reproducible.
kmeans = KMeans(n_clusters= 3 , random_state= 2 )
# --------------------------------------------------------------------------------------------------

#   Fit the Model (Only on Churned Customers)
#  we're clustering only churned customers (Churn == 1) based on:
#  Tenure_norm
#  MonthlyCharges_norm
#  These are the standardized values of how long they’ve been with the company and how much they pay.
kmeans.fit(df_cal[df_cal.Churn==1][['Tenure_norm','MonthlyCharges_norm']])
# --------------------------------------------------------------------------------------------------


#  Create Colors for Each Cluster
#  kmeans.labels_ contains the cluster assigned to each churned customer.

#  This line assigns:

#  Green to cluster 0

#  Blue to cluster 1

#  Red to cluster 2
kmeans_colors = ['green' if c == 0 else 'blue' if c==1 else 'red' for c in kmeans.labels_ ] 
# --------------------------------------------------------------------------------------------------

#  Assign Cluster Labels to the Dataset
#  This assigns the predicted cluster label to the churned customers in the Cluster column.
#  we're using .fit_predict() to fit and return the cluster labels in one step.

df_cal.loc[(df_cal.Churn == 1) ,'Cluster'] = kmeans.fit_predict(df_cal[df_cal.Churn == 1 ][['Tenure_norm' , 'MonthlyCharges_norm']])
# --------------------------------------------------------------------------------------------------

#Plotting the Clusters
fig= plt.figure(figsize=(12,8))

#  Scatter Plot the Data Points (Colored by Cluster)
#  Scatter plot for churned customers
#  X-axis: Tenure_norm, Y-axis: MonthlyCharges_norm
#  Color: based on cluster assignment (green, blue, or red)
#  alpha=0.25: makes points transparent so overlapping points are visible

plt.scatter(x= 'Tenure_norm' , y= 'MonthlyCharges_norm' ,data=df_cal[df_cal.Churn==1], color= kmeans_colors , alpha= 0.25)

#  Plot the Cluster Centers
plt.scatter(x = kmeans.cluster_centers_[:,0] , y= kmeans.cluster_centers_[:,1] , color= 'black' ,marker= 'X' , s= 100 )

# --------------------------------------------------------------------------------------------------
#   Final Touches  
plt.xlabel("Tenure")
plt.ylabel("Monthly Charges")
plt.title("Tenure vs Monthly charges : Churn customer cluster", fontsize=15)
plt.show()

```

```python
# distribution of cluster
print(df_cal[df_cal['Churn']==1].Cluster.value_counts(normalize=True))
# Calculate normalized value counts for churned customers

cluster_dis = df_cal[df_cal['Churn'] == 1]['Cluster'].value_counts(normalize = True)
# Plot using Seaborn
sns.barplot(hue=cluster_dis.index,  # Cluster numbers (0, 1, 2)
            y=cluster_dis.values,
            palette=['g', 'r', 'b'],
            alpha=0.6)
plt.xlabel("Clusters")
plt.ylabel("Distribution")
plt.title("Distribution of Clusters")
plt.show()
```

Let's analyse characteristics of different clusters based on information available in dataset.

```python
def cluster_analysis(df,cluster_gp,feature):
    group_gp = cluster_gp[feature].value_counts(normalize=True).to_frame()

    #pd.concat([group_gp.index.name , group_gp.values])
    group_gp.columns = ['Percentage']
    group_gp = group_gp.reset_index()
    group_gp_new = group_gp.copy()  
    return (df.groupby('Churn')[feature].value_counts(normalize=True), group_gp_new)
```

```python
(gender_dist , gender_cluster_cnt) = cluster_analysis(df_cal,df_cluster_gp,'Gender')
print(gender_dist)
print(gender_cluster_cnt)
```

This shows the **gender distribution within each churned cluster** (Clusters 0, 1, and 2 among churned customers):

- **Cluster 0**: 53.5% male, 46.5% female

- **Cluster 2**: 53.2% female, 46.8% male

👉 **Interpretation**:

- There are **slight imbalances in gender** per cluster.
- Cluster 0 and 1 have **more males**, while Cluster 2 has **more females**.
- These differences are **not huge**, but might hint at **slightly different churn behavior patterns** by gender across clusters.
